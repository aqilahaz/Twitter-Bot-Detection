```python
import pandas as pd
import datetime
import numpy as np
import tweepy
from time import sleep
import csv
from datetime import datetime, date, time, timedelta
import re
import numpy as np


def preprocess(df):
        df = pd.DataFrame(df)

        df.drop(['tweet_lat',
                        'tweet_long',
                        'tweet_in_reply_to_screen_name',
                        'tweet_direct_reply',
                        'tweet_hashtags',
                        'tweet_urls',
                        'tweet_user_mentions',
                        'tweet_media_type',
                        'tweet_contributors'], 
                        axis=1, 
                        inplace=True)
        df = df.dropna( subset = ['tweet_text'])
        train = df[['name',
                        'username',
                        'tweet_created_at',
                        'tweet_year',
                        'tweet_month',
                        'tweet_day',
                        'tweet_hour',
                        'tweet_text',
                        'tweet_source',
                        'tweet_retweet_status',
                        'tweet_retweet_count',
                        'tweet_favorite_count',
                        'tweet_hashtags_count',
                        'tweet_urls_count',
                        'tweet_user_mentions_count']]


        train3 = df[['name','username',
                        'followers_count',
                        'listed_count',
                        'friends_count',
                        'favorites',
                        'created_at',
                        'verified',
                        'default_profile',
                        'default_profile_image',
                        'location',
                        'statuses_count',
                        'description',
                        'url',
                        'geo_enabled']]
        train['word_count'] = train['tweet_text'].apply(lambda x: len(str(x).split(" "))) #count the word
        train['char_count'] = train['tweet_text'].str.len() 
        #measure the lexical diversity of a word


        def avg_word(sentence):
                words = sentence.split()
                return (sum(len(word) for word in words)/len(words))


        train['avg_word'] = train['tweet_text'].apply(lambda x: avg_word(x))


        def lexical_diversity(tweet_text):
                if len(tweet_text) == 0:
                        diversity = 0
                else: 
                        diversity = float(len(set(tweet_text))) / len(tweet_text)
                return diversity

        train['lexical'] = train['tweet_text'].apply(lambda x: lexical_diversity(x))

        train['tweet_created_at'] =  pd.to_datetime(train['tweet_created_at']) #change datatype of tweet_created_column
        train.sort_values(["username", "tweet_created_at"]) #sort values based on username and time
        train['time_diff'] = (train['tweet_created_at'] - train['tweet_created_at'].shift()).abs()  / np.timedelta64(1, 'h')
        train['time_diff'] = train['time_diff'].fillna(0) #fill missing value in time_diff with zero
        g_sum = 50 #there is 50 tweet per user extracted, so n=50
        values = train['time_diff']/g_sum #measure the probability for this interval
        train['Entropy'] = -(values*np.log(values)) #time interval entropy
        train['Entropy']=train['Entropy'].fillna(0) #fill missing value in entropy column with zero
        train.drop(train[train.Entropy < 0].index, inplace=True) #if value is negative, it will be eliminated. The reason is because it was an outlier
        grouped = train.groupby("username")
        groupie = grouped.agg(np.mean)
        groupie.drop(['tweet_year','tweet_month', 'tweet_day','tweet_hour','time_diff'], axis=1, inplace=True)
        groupie.to_csv('sample.csv')
        

        train3=train3.drop_duplicates(keep='first', inplace=False)
        train3['created_at']= pd.to_datetime(train3['created_at'])
        now = pd.Timestamp('now')
        train3['age_in_days'] = (now - train3['created_at']).dt.days
        train3["created_at"]= pd.to_datetime(train3["created_at"]) 
        train3['ratio statuses_count_per_age']=train3['statuses_count']/train3['age_in_days']
        #Number of favourites/age
        train3['ratio favorites_per_age']=train3['favorites']/train3['age_in_days']
        #Ratio friends/follower
        train3['ratio_friends_per_followers']=train3['friends_count']/train3['followers_count']
        train3['description'].fillna('x', inplace = True)
        train3['length_of_bio']=train3['description'].str.len()
        train3['reputation']= train3['followers_count']/(train3['followers_count']+ train3['friends_count'])
        train3.drop(['geo_enabled','created_at'], axis=1,inplace=True)
        train3['url'] = pd.notnull(train3['url']) 
        train3['location']=pd.notnull(train3['location'])
        train3['contains_bot_name']=train3['description'].str.contains("\b(bot|b0t|updates|hourly|automatically|generating|generated|every|computer-generated|twitterbot|automated|FakeBots|')\b|Bots", 
                                                                flags=re.IGNORECASE, 
                                                                regex=True)
        train3['contains_bot_name'].fillna(0).astype(bool).sum(axis=0)
        train3.drop(['name','description'], inplace=True, axis=1)
        train3["length_of_bio"].fillna(0, inplace = True) 
        train3["reputation"].fillna(0, inplace = True) 
        train3["contains_bot_name"].fillna(False, inplace=True)
        train3.replace(np.inf, 0)
        train3.replace(-np.inf, 0)
        #join two file

        group=pd.read_csv('sample.csv')
        group.set_index('username')
        train3.set_index('username')
        group=group.merge(train3,how='left')


        group=group.replace([np.inf, -np.inf], np.nan)
        group=group.replace([np.inf, -np.inf], np.nan).dropna(how="all")
        group=group.drop(['username'],axis=1)
        group=group[['Entropy',
                        'age_in_days',
                        'avg_word',
                        'char_count',
                        'contains_bot_name',
                        'default_profile',
                        'default_profile_image',
                        'favorites',
                        'followers_count',
                        'friends_count',
                        'length_of_bio',
                        'lexical',
                        'listed_count',
                        'location',
                        'ratio favorites_per_age',
                        'ratio statuses_count_per_age',
                        'ratio_friends_per_followers',
                        'reputation',
                        'statuses_count',
                        'tweet_favorite_count',
                        'tweet_hashtags_count',
                        'tweet_retweet_count',
                        'tweet_retweet_status',
                        'tweet_urls_count',
                        'tweet_user_mentions_count',
                        'url',
                        'verified',
                        'word_count']]
        for_index=df[['username','name',
                        'followers_count',
                        'friends_count',
                        'statuses_count']]
        for_index=for_index.drop_duplicates(keep='first', inplace=False)
        for_index.to_csv("user.csv",index=False)
        pr=group.values.tolist()
        return pr


```

```python
df=pd.read_csv('coba.csv')
preprocess(df)
```
